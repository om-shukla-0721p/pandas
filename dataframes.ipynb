1. dataframe -> a tabular data structure with rows and col
<br> it is 2D similar to excel spreadsheet


```python
import pandas as pd
data={"name":["Spongebob","Patrick","Squidward"],
      "age": [30,35,50]}

df=pd.DataFrame(data,index=["a","b","c"])
print(df)
print()
print(df.loc["a"])
print()
print(df.iloc[2])
```

2. adding new cols and rows

```python
# directly
# add a new col
df["job"]=["cook","N/A","cashier"]
print(df)
print()

# add a new row
new_row=pd.DataFrame([{"name":"sandy",
                      "age":28,
                      "job":"engineer"},
                      
                      {"name":"eugene",
                      "age":60,
                      "job":"manager"}],index=["employee 4","employee 5"])
df=pd.concat([df,new_row])
print(df)
print()
```
