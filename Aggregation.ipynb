1) Aggregate func== reduce a set of values into a single summary value 
2) used to separate and analyze data 
3) often used with the groupby() function

1.1) whole Dataframe

```python
import pandas as pd
df=pd.read_csv("pokemon_data_for_learning/pokemon_gen_1_data.csv")
# print(df.mean(numeric_only=True))
# print(df.sum(numeric_only=True))
# print(df.min(numeric_only=True))
# print(df.max(numeric_only=True))
print(df.count())
```

1.2) single column

```python
print(df["Height"].mean())
print(df["Weight"].min())
print(df["Height"].sum())
print(df["Legendary"].count())
print(df["Height"].max())
```

2.3) groupby()

```python
group=df.groupby("Type1")
# Height thing
# print(group["Height"].mean())
# print(group["Height"].min())
# print(group["Height"].max())
# print(group["Height"].sum())
# print(group["Height"].count())

# Weight thing

print(group["Weight"].mean())
print(group["Weight"].min())
print(group["Weight"].max())
print(group["Weight"].sum())
print(group["Weight"].count())
```
