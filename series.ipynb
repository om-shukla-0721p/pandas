1. series->pandas 1 dimensional labeled array 

```python
import pandas as pd
data=[100.0,102,103.0]
series=pd.Series(data,index=["a","b","c"])
series.loc["c"]=200
print(series.loc["c"])
print(series.iloc[1])
print(series.iloc[2])
```

```python
data=[100,102,108,200,240]
series=pd.Series(data,index=["a","b","c","d","e"])

print(series[series<200])
```

```python
calories={"day1":1750,
          "day2":2150,
          "day3":1701}
series=pd.Series(calories)
print(series)
print(series.iloc[2])
print(series[series>2000])
print(series>2000)
```
